```python
params = {'legend.fontsize': 20,
          'figure.figsize': (9.5, 8.5),
          'axes.labelsize': 27,
          'axes.titlesize': 27,
          'xtick.labelsize': 27,
          'ytick.labelsize': 27,
          'image.cmap': 'jet',
          'lines.linewidth': 3,
          'lines.markersize': 5,
          'font.family': 'sans-serif'}

plt.rc('text', usetex=False)
plt.rc('font', family='serif')
plt.rcParams.update(params)
```

```python
studies_list = ['kickerSPSwake_2020_oldMKP.wake', 'newkickers_Q26_2018.txt', 'newkickers_Q26_2018_modified.txt' ]
labels = ['Q20', 'Q26, original', 'Q26, modified']
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15,10))
fig.suptitle('Wakefields SPS', fontsize=27)

linestyles = ['-', '-', '--']

for index, study in enumerate(studies_list):
    n_turns_wake = 1 # for the moment we consider that the wakefield decays after 1 turn
    wakefile1 = (study)
    ww1 = WakeTable(wakefile1, ['time', 'dipole_x', 'dipole_y', 'quadrupole_x', 'quadrupole_y'], n_turns_wake=n_turns_wake)

    time = ww1.wake_table['time']
    dipole_x = ww1.wake_table['dipole_x']
    dipole_y = ww1.wake_table['dipole_y']
    quadrupole_x = ww1.wake_table['quadrupole_x']
    quadrupole_y = ww1.wake_table['quadrupole_y']
    
    print(f'Kicks {len(time)}')
    print(time[-1])

    
    ax1.plot(time, dipole_x, linestyle=linestyles[index], label=f'{labels[index]}')
    ax2.plot(time, dipole_y, linestyle=linestyles[index])
    ax3.plot(time, quadrupole_x, linestyle=linestyles[index])
    ax4.plot(time, quadrupole_y, linestyle=linestyles[index])


# Limit axis
turns_max = 10
ax1.set_xlim(-2, turns_max)
ax2.set_xlim(-2, turns_max)
ax3.set_xlim(-2, turns_max)
ax4.set_xlim(-2, turns_max)

ax1.set_ylim(-2, 15)
ax2.set_ylim(-2, 15)
ax3.set_ylim(-12, 10)
ax4.set_ylim(-12, 10)

# Grids
ax1.grid(linestyle='dashed')
ax2.grid(linestyle='dashed')
ax3.grid(linestyle='dashed')
ax4.grid(linestyle='dashed')

# Axis labels
ax1.set_ylabel('Dipole-x \n [V/pC/mm]')
ax2.set_ylabel('Dipole-y \n [V/pC/mm]')
ax3.set_ylabel('Quadrupole-x \n [V/pC/mm]')
ax4.set_ylabel('Quadrupole-y \n [V/pC/mm]')
ax1.set_xlabel('Waketime (ns)')
ax2.set_xlabel('Waketime (ns)')
ax3.set_xlabel('Waketime (ns)')
ax4.set_xlabel('Waketime (ns)')

ax1.legend()
plt.subplots_adjust(wspace=0.4, hspace=0.4)


#plt.tight_layout()

savefig = False
if savefig:
    plt.savefig(f'./Q26_wakes_WaketimeMax{turns_max}ns.png')
```
